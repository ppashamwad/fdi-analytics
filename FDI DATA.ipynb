```python
#craeting Average Exchange Rate list  :-Reference (RBI website)
Rates =[45,68,47,69,48,39,45,95,44,93,44,27,45,24,40,26,45,99,47,44,45,56,47,92,54,40,60,50,61,14,65,46,67,7]
```

Converting American $ toIndian ₹:

```python
#Cteating a function to convert FDI'S value from USD to INR
def multiply_columns(df, col_list,num):
    for col in col_list:
        df[col] =df[col] * Rates[col_list.index(col)]/10
        return df
```

```python
FDI_InUSD=FDI.copy()
FDI_02 = multiply_columns(FDI, Year, Rates)
```

```python
#FDI INFLOWS (amount in ₹ crores)
FDI_02.style.set_caption("FDI INFLOWS (Amount in ₹ crores)").format(precision=2)
```

Unpivioting/Melting/reshaping Dataframes from wide to long format

```python
#Unpivoting melt Dataframe
melt = pd.melt(FDI_InUSD, Sectors, value_vars = Year, var_name='Year', value_name='FDI(US$ Million)',ignore_index=True)
melt1=round(melt,2)
melt
```

```python
#Unpivoting melt01 Datframe
melt01 = pd.melt(FDI_02, id_vars = Sectors, value_vars = Year, var_name='Year', value_name='FDI(₹ Crores)',ignore_index=True)
melt01=round(melt01,2)
melt01
            
```

```python
#Merging the FDI(US$ Million) column of melt Dataframe into melt01Dataframe
Merged=melt01.merge(melt,how='left')
Merged
```

```python
#sorting the sectors and year columns
Sorted = Merged.sort_values(['Sector','Year'], ignore_index=True)
Sorted
```

```python
#repla cing someLong Valuesof sector column to short form
Sorted = Sorted[['Sector', 'FDI(₹ Crores)', 'FDI(US$ Million)',]].replace(["CONSTRUCTION DEVELOPMENT: TownshipS, housing, built-up infrastruture and construction-development project","SERVICES SECTOR(Fin.Banking,Insurance,Non Fin/Business,Outsorcing,R&D,Courier,Tech. Testing and Analysis,other)",
  'TEA AND COFFEE(PROCESSING & WAREHOUSING COFFEE & RUBBER)'],["CONSTRUCTION DEVELOPMENT","SERVICES SECTOR",'TEA AND COFFEE'])                                                                     
                                        
```

Sectorwise Total FDI 2000-17

```python
#Grouping by Sector Column to find total FDI Inflow per sector from FY2000-01 to FY2016-17
Sectorwise_fdi = Sorted.groupby('Sector').sum()
Sectorwise_fdi.sort_values(by='FDI(US$ Million)',ascending=False)
```
